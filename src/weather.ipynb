```python
#   Mesmo passo para PORTO VELHO
from pathlib import Path

pvelho_path = Path(r'c:\Users\guica\OneDrive\Documentos\GitHub\CurrentFlow-2.0\data\raw\inmet\dados_porto_velho_A925_D.csv')
pvelho_df = pd.read_csv(pvelho_path, sep=';', skiprows=10)

pvelho_df = pd.DataFrame(pvelho_df[['Data Medicao', 'PRECIPITACAO TOTAL, DIARIO (AUT)(mm)', 'TEMPERATURA MEDIA, DIARIA (AUT)(°C)']])
pvelho_df['Data Medicao'] = pd.to_datetime(pvelho_df['Data Medicao'])
pvelho_df.isnull().any()
```

```python
pvelho_df = pvelho_df.fillna(0)
pvelho_df.isnull().any()
```

```python
# Norte

def padronizar(df):
    df = df.rename(columns={
        'Data Medicao': 'data',
        'PRECIPITACAO TOTAL, DIARIO (AUT)(mm)': 'precipitacao',
        'TEMPERATURA MEDIA, DIARIA (AUT)(°C)': 'temperatura'
    })
    df['data'] = pd.to_datetime(df['data'])
    return df

pvelho_df = padronizar(pvelho_df)
belem_df  = padronizar(belem_df)
manaus_df = padronizar(manaus_df)
```

```python
df_norte = pd.concat([pvelho_df, belem_df, manaus_df], ignore_index=True)
# print(df_norte.columns)

def comma_replace(df):
    df['temperatura'] = df['temperatura'].astype(str).str.replace(',', '.').astype(float)
    df['precipitacao'] = df['precipitacao'].astype(str).str.replace(',', '.').astype(float)
    return df

df_norte = comma_replace(df_norte)

df_norte = pd.DataFrame(df_norte.groupby('data')[['precipitacao', 'temperatura']].mean().reset_index())

df_norte
```
