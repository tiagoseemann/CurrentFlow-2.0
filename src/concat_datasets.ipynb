```python
dfs_weather_sul = []

for path in REGIONS['sul']:
    df = load_inmet(path=path)
    df = select_columns(df=df)
    df = standardize(df=df)
    df = fix_commas(df=df)
    dfs_weather_sul.append(df)

weather_sul = merge_region(dfs_weather_sul)
print(weather_sul)
```
